```python
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind
from collections import defaultdict
from IPython.display import display

# Assuming `data` is your DataFrame containing the relevant columns
data_df = data.copy()
data_df['state'] = data['state']
data_df['presence_absence'] = data['presence_absence']

# Extract the list of _raw variables
raw_variables = [col for col in data.columns if '_raw' in col]

# Dictionary to store counts
state_counts = defaultdict(lambda: {'presence_absence_0': 0, 'presence_absence_1': 0})

# Initialize the balanced data DataFrame
balanced_data = pd.DataFrame()

# Group the data by 'state'
for state, group in data_df.groupby('state'):
    # Separate the group into presence_absence 0 and 1
    group_0 = group[group['presence_absence'] == 0]
    group_1 = group[group['presence_absence'] == 1]
    
    # Store counts before balancing
    state_counts[state]['presence_absence_0'] = len(group_0)
    state_counts[state]['presence_absence_1'] = len(group_1)
    
    # Determine the number of samples to select (the minimum of the two groups)
    num_samples = min(len(group_0), len(group_1))
    
    # Randomly sample from both groups to balance them
    if num_samples > 0:
        balanced_group_0 = group_0.sample(num_samples, random_state=42)
        balanced_group_1 = group_1.sample(num_samples, random_state=42)
    
        # Concatenate the balanced samples and add them to the balanced_data DataFrame
        balanced_state_group = pd.concat([balanced_group_0, balanced_group_1])
        balanced_data = pd.concat([balanced_data, balanced_state_group])

# Remove duplicates
balanced_data = balanced_data.drop_duplicates()
balanced_data
```

```python

# Initialize a dictionary to store test results
test_results = {}

# Perform tests for each _raw variable
for var in raw_variables:
    # Group the data by presence_absence
    group_0 = balanced_data[balanced_data['presence_absence'] == 0][var]
    group_1 = balanced_data[balanced_data['presence_absence'] == 1][var]
    
    # Perform a t-test to compare the means of the two groups
    t_stat, p_value = ttest_ind(group_0, group_1)
    
    # Store results in the dictionary
    test_results[var] = {
        't-Statistic': t_stat,
        'p-value': p_value,
        'Group 0 Mean': group_0.mean(),
        'Group 1 Mean': group_1.mean()
    }
    
    # Print results for each variable
    print(f"Results for {var}:")
    print("t-Statistic:", t_stat)
    print("p-value:", p_value)
    print("Group 0 Mean:", group_0.mean())
    print("Group 1 Mean:", group_1.mean())
    print("\n")

# Convert the test results to a DataFrame for better visualization
test_results_df = pd.DataFrame(test_results).T
display(test_results_df)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
from collections import defaultdict
from IPython.display import display

# Assuming `data` is your DataFrame containing the relevant columns
data_df = data.copy()
data_df['state'] = data['state']
data_df['presence_absence'] = data['presence_absence']

# Extract the list of _raw variables
raw_variables = [col for col in data.columns if '_raw' in col]

# Dictionary to store chi-square results
chi_square_results = defaultdict(list)

# Perform Chi-Square Test for each _raw variable grouped by state
for state, state_group in data_df.groupby('state'):
    for var in raw_variables:
        # Separate the group into presence_absence 0 and 1
        group_0 = state_group[state_group['presence_absence'] == 0]
        group_1 = state_group[state_group['presence_absence'] == 1]
        
        # Determine the number of samples to select (the minimum of the two groups)
        num_samples = min(len(group_0), len(group_1))
        
        # Randomly sample from both groups to balance them
        if num_samples > 0:
            balanced_group_0 = group_0.sample(num_samples, random_state=42)
            balanced_group_1 = group_1.sample(num_samples, random_state=42)
            
            # Concatenate the balanced samples
            balanced_state_group = pd.concat([balanced_group_0, balanced_group_1])
            
            # Create a contingency table for the variable and presence_absence
            contingency_table = pd.crosstab(balanced_state_group['presence_absence'], balanced_state_group[var])
            
            if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1:  # Ensure the table is not empty
                chi2, p_value, dof, expected = chi2_contingency(contingency_table)
                
                # Store results in the dictionary
                chi_square_results[var].append({
                    'State': state,
                    'Chi-Square Statistic': chi2,
                    'p-value': p_value,
                    'Degrees of Freedom': dof,
                    'Expected Frequencies': expected
                })
                # Print results for each variable in each state
                print(f"Results for {var} in {state}:")
                print("Contingency Table:\n", contingency_table)
                print("Chi-Square Statistic:", chi2)
                print("p-value:", p_value)
                print("Degrees of Freedom:", dof)
                print("Expected Frequencies:\n", expected)
                print("\n")
            else:
                print(f"Not enough data to perform chi-square test for {var} in state {state}.\n")

# Convert the chi-square results to a DataFrame
chi_square_results_df = pd.DataFrame([
    {
        'Variable': var,
        'State': res['State'],
        'Chi-Square Statistic': res['Chi-Square Statistic'],
        'p-value': res['p-value'],
        'Degrees of Freedom': res['Degrees of Freedom'],
        'Expected Frequencies': res['Expected Frequencies']
    }
    for var, results in chi_square_results.items()
    for res in results
])

# Display the results in a DataFrame
display(chi_square_results_df)

# Save the aggregated results to a CSV file
chi_square_results_df.to_csv('../data/chi_square_results.csv', index=False)
```

## KMeans Clustering Analysis

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
from sklearn.metrics import silhouette_score

# Scaling the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[variables_to_plot])

# Initialize lists to store the results
wcss = []
silhouette_scores = []

# Compute WCSS and Silhouette Scores for each number of clusters from 2 to 10
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_data, cluster_labels))

# Plotting both WCSS and Silhouette Scores side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# WCSS Plot
ax1.set_title('Elbow Method (WCSS)')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('WCSS')
ax1.plot(range(2, 11), wcss, marker='o', color='tab:red')
ax1.grid(True)

# Silhouette Score Plot
ax2.set_title('Silhouette Scores')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Silhouette Score')
ax2.plot(range(2, 11), silhouette_scores, marker='o', color='tab:blue')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

Choosing 6 clusters provides a good balance between minimizing WCSS and maximizing silhouette score.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Assuming 'data_with_clusters' contains all the necessary environmental metrics and the 'presence_absence' label
X = data_with_clusters[variables_to_plot]  # Independent variables from cluster analysis
y = data_with_clusters['presence_absence']  # Dependent variable indicating industry impact

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # 70% training and 30% testing

# Creating a Logistic Regression model
logistic_model = LogisticRegression(max_iter=1000)  # Increased max_iter for convergence in case of non-separation
logistic_model.fit(X_train, y_train)  # Fitting the model to the training data

# Making predictions on the test set
y_pred = logistic_model.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Output the results
print("Accuracy of the model:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)
```
